```python
import pandas as pd 
import numpy as np
import matplotlib
import seaborn as sns
sns.set_style('whitegrid', {'axes.grid' : False})
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn
import datetime
```

```python
data = pd.read_csv('Alameda_County_Sheriff_Crime_Reports.csv')
data
```

```python
for i in data['DateTime'].index:
    data.loc[i, 'year'] = datetime.datetime.strptime(data['DateTime'].loc[i], '%m/%d/%Y %H:%M:%S %p').year
```

```python
data.groupby('year').count()
#crime data looks like it can be compared across 2012 to 2016
```

```python
create_df = pd.DataFrame(data={'2012': data[data['year'] == 2012].groupby('CrimeDescription').count()['DateTime'],\
                               '2013': data[data['year'] == 2013].groupby('CrimeDescription').count()['DateTime'],\
                               '2014': data[data['year'] == 2014].groupby('CrimeDescription').count()['DateTime'],\
                               '2015': data[data['year'] == 2015].groupby('CrimeDescription').count()['DateTime'],\
                               '2016': data[data['year'] == 2016].groupby('CrimeDescription').count()['DateTime']
                              })
create_df
```

Although the descriptions give a better sense of what's going on during the call, the codes are probably a better way to classify the calls to give better numbers for comparison from 2012 to 2016/2017. Let's take a look at the codes and what they mean. There are 57 different codes rather than 913 different descriptions. We'll groupby the codes for now, but when investigating one specific type of call, "72 Hour Mental Health", we'll be pulling from the descriptions. 

```python
codes = pd.read_csv('ac_crime_categories.csv')
codes
```

```python
create_codes_df = pd.DataFrame(data={'2012': data[data['year'] == 2012].groupby('CrimeCode').count()['DateTime'],\
                               '2013': data[data['year'] == 2013].groupby('CrimeCode').count()['DateTime'],\
                               '2014': data[data['year'] == 2014].groupby('CrimeCode').count()['DateTime'],\
                               '2015': data[data['year'] == 2015].groupby('CrimeCode').count()['DateTime'],\
                               '2016': data[data['year'] == 2016].groupby('CrimeCode').count()['DateTime']
                              })
create_codes_df['avg'] = create_codes_df.mean(axis=1)
create_codes_df
```

Let's visualize some of these crimes! We'll look at ten. 

```python
create_codes_df[create_codes_df['avg'] > 500][:10].T.plot(kind='line', marker='o')
plt.ylabel('Number of Calls', fontsize=14)
plt.ylim(0, 3000)
labels=['13B: Simple Assault', '220: Burglary', '240: Motor Vehicle Theft', '35A: Drug Offense', '35B: Drug Equipment',\
        '90C: Disorderly Conduct', '90D: DUI', '90Z: All Other', '999: Informational']
plt.legend(labels, bbox_to_anchor=(1.1, 1.1))
```
