```python
def extract_weekday(table):
    for i in table['DateTime'].index:
        table.loc[i, 'weekday'] = datetime.datetime.strptime(table['DateTime'].loc[i], '%m/%d/%Y %H:%M:%S %p').weekday()
    return table
```

```python
temp = extract_weekday(data)
```

```python
def extract_hour(table): 
    for i in table['DateTime'].index:
        table.loc[i, 'hour'] = datetime.datetime.strptime(table['DateTime'].loc[i], '%m/%d/%Y %I:%M:%S %p').hour
    return table
```

```python
final = extract_hour(temp)
```

## Looking at Crime during the Week

```python
final.groupby(['weekday', 'hour']).count()['DateTime']
_data = pd.DataFrame({'count' : final.groupby(['weekday', 'hour']).size()}).reset_index()
_heatmap = _data.pivot('weekday', 'hour', 'count')
ax = sns.heatmap(mh_heatmap)
ax.set_yticklabels(['Sun', 'Sat', 'Fri', 'Thu', 'Wed', 'Tue','Mon'], fontsize=10)
```

## Looking at specific types of crimes

Crimes Related to the 72 Hour Mental Health Hold 

```python
MH_data = pd.DataFrame({'count' : final[final['CrimeDescription'] == '72 HOUR MENTAL HEALTH'].groupby(['weekday', 'hour']).size()}).reset_index()
mh_heatmap = MH_data.pivot('weekday', 'hour', 'count')
ax = sns.heatmap(mh_heatmap)
ax.set_yticklabels(['Sun', 'Sat', 'Fri', 'Thu', 'Wed', 'Tue','Mon'], fontsize=10)
```

```python
def create_heatmap(code):
    _data = pd.DataFrame({'count' : final[final['CrimeCode'] == code].groupby(['weekday', 'hour']).size()}).reset_index()
    _heatmap = _data.pivot('weekday', 'hour', 'count')
    ax = sns.heatmap(_heatmap)
    ax.set_yticklabels(['Sun', 'Sat', 'Fri', 'Thu', 'Wed', 'Tue','Mon'], fontsize=10)
    title = codes[codes['Federal IBR#'] == code]['Federal Description']
    plt.title(title)
    plt.show()
#currently does not plot the nan values 
```

```python
for i in create_codes_df.index:
    print(i)
    create_heatmap(i)
```
